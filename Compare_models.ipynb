```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Need to fit first before accessing params
X, y = make_classification(n_samples=100, n_features=4, random_state=42)
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

clf = MLPClassifier(hidden_layer_sizes=(3,), activation="tanh", 
                    solver='lbfgs', random_state=1, max_iter=8000)
clf.fit(X_scaled, y)

# Print weights and biases
total_params = 0
for i, (weights, biases) in enumerate(zip(clf.coefs_, clf.intercepts_)):
    layer_params = weights.size + biases.size
    total_params += layer_params
    print(f"Layer {i+1}:")
    print(f"  Weights shape: {weights.shape} = {weights.size} params")
    print(f"  Biases shape:  {biases.shape} = {biases.size} params")
    print(f"  Layer total:   {layer_params} params")

print(f"\nTotal trainable parameters: {total_params}")
```
