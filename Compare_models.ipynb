```python
from matplotlib import pyplot as plt
import pandas as pd
from MLP_classifier import MLP_classifier
from VQC_classifier import VQC_classifier
from preprocess_data import preprocess, preprocess_mnist_digits
import numpy as np
from sklearn.datasets import make_classification
from sklearn.preprocessing import LabelEncoder

# distributions = [
#     ("raw", None),
#     ("scaled", None),
#     ("pca", 4),
#     ("pca", 6),
# ]

# df = pd.read_csv("/Users/nataliaahsan/qml/heart.csv")

# # Encode categorical columns
# le = LabelEncoder()
# for col in df.select_dtypes(include='object').columns:
#       df[col] = le.fit_transform(df[col])

# # turns the data into arrays
# X = df.drop(columns=["HeartDisease"]).values
# y = df["HeartDisease"].values

X, y = make_classification(n_samples=1000, n_features=4, n_informative=4, n_redundant = 0, n_classes=2, random_state=42, class_sep=0.1)

# X, y = preprocess_mnist_digits(digit1=3, digit2=8, n_samples=1000)

print(f"No. of features: {X.shape[0]}")

classical_model = MLP_classifier()
classical_model.fit_predict_evaluate(X, y, preprocess_mode='scaled')
classical_accuracy = classical_model.get_accuracies()
classical_mean = np.mean(classical_accuracy)
classical_std = np.std(classical_accuracy)
classical_times = classical_model.get_train_times()
print("MLP mean training time:", np.mean(classical_times))
print("MLP mean accuracy:", classical_mean)
print("MLP standard deviation:", classical_std)

quantum_model = VQC_classifier()
quantum_model.fit_predict_evaluate(X, y, preprocess_mode='scaled')
quantum_accuracy = quantum_model.get_accuracies()
quantum_mean = np.mean(quantum_accuracy)
quantum_std = np.std(quantum_accuracy)
quantum_times = quantum_model.get_train_times()
print("VQC mean training time:", np.mean(quantum_times))
print("VQC mean accuracy:", quantum_mean)
print("VQC standard deviation:", quantum_std)

for i, fold_obj in enumerate(quantum_model.get_objective_values()):                                                                                            
      plt.plot(fold_obj, label=f"Fold {i+1}")                                                                                                                    
plt.xlabel("Iteration")                                                                                                                                        
plt.ylabel("Objective Value")                                                                                                                                  
plt.title("VQC Loss Curve - class sep 0.1")
plt.legend()
plt.show()

# plt.figure(figsize=(12, 6))
# plt.title("MLP Metrics Graph")
# plt.plot(classical_accuracy, label="Classical - Accuracy Curve")
# plt.plot(quantum_accuracy, label="Quantum - Accuracy Curve")
# plt.xlabel("Fold")
# plt.ylabel("Metric Value")
# plt.legend()
# plt.gcf().savefig(
#     "mlp_vs_vqc_raw_accuracy_randomData_reps=2_hidden_layers=1.png",
#     dpi=300,
#     bbox_inches="tight"
# )
# plt.show()
```
